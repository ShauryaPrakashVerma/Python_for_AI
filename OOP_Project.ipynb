```python
import nlpcloud

class NLPApp:

  def __init__(self):
    self.__database = {}
    self.__first_menu()
    # print(self.__database)  -->  prints at last, because first self.__menu() is executed completely.

  def __first_menu(self):

    choice_1 = input("""
    Hello! How would you like to proceed?
    1. Press 1 to register.
    2. Press 2 to login.
    3. Press 3 to exit.
    """)

    if choice_1 == "1":
      self.__register()
    elif choice_1 == "2":
      self.__login()
    else:
      exit()

  def __register(self):
    mail = input("Enter email ID: ")
    if mail in self.__database:
      print("Mail already registered. Please, try with a different mail ID.")
      self.__first_menu()
    else:
      username = input("Enter username: ")
      password = input("Enter password: ")
      self.__database[mail] = [username, password]
      print("Registration successfull. Now login")
      self.__first_menu()


  def __login(self):
    mail = input("Enter email ID: ")
    password = input("Enter password: ")
    if mail in self.__database:
      if self.__database[mail][1] == password:
        print("Login successfull!")
        self.__second_menu()
      else:
        print("Wrong Password. Try again")
        self.__login()
    else:
      print("This email is not registered")

  def __second_menu(self):
    choice_2 = input("""
    1. Enter 1 for NER
    2. Enter 2 for Language Detection
    3. Enter 3 for Sentiment analysis
    4. Enter anything else for logout.
    """)

    if choice_2 == "1":
      self.__ner()
    elif choice_2 == "2":
      self.__language_detection()
    elif choice_2 == "3":
      self.__sentiment_analysis()
    else:
      print("Logout")

  def __ner(self):
    para = input("Enter the paragraph: ")
    searched_term = input("Enter the term: ")
    client = nlpcloud.Client("gpt-oss-120b", "c622cb450d836ee37e4561658bf3ec7b5939da0b", gpu=True)
    response = client.entities(
        para,
        searched_entity = searched_term
    )

    for i in response["entities"]:
      print(i["text"])

    self.__second_menu()


  def __language_detection(self):
    para = input("Enter the paragraph: ")
    client = nlpcloud.Client("python-langdetect", "c622cb450d836ee37e4561658bf3ec7b5939da0b", gpu=False)
    response = client.lang_detection(
        para
    )

    print(response)

  def __sentiment_analysis():
    para = input("Enter the paragraph: ")
    client = nlpcloud.Client("gpt-oss-120b", "c622cb450d836ee37e4561658bf3ec7b5939da0b", gpu=True)
    response = client.sentiment(
        """I hate convoluted products with poor documentation. I see such products all the time. But NLP Cloud is not like that!""",
        target="NLP Cloud"
    )

    print(response)


s = NLPApp()
```

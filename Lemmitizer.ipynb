```python
import nltk
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt_tab')
```

```python
nltk.download('wordnet')
nltk.download('omw')
```

```python
text = "The taller were happier and more energetic than their older, dullest classmates."
tokens = word_tokenize(text)
print(tokens)
```

```python
for i in tokens:
  print(WordNetLemmatizer().lemmatize(i))
  print(WordNetLemmatizer().lemmatize(i,pos='a'))
```

```python
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# multiple line plots in same graph
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
```

```python
# select colors of the line graphs
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'red')
plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
```

```python
# lines style, linewidth
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle='dashdot')   #dashed
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'red', linestyle='dotted', linewidth=3)
plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
```

```python
# visualize the dots
# marker, markersize
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle='dashdot',marker='D', markersize=10)   #dashed
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'red', linestyle='dotted', linewidth=3, marker='o')
plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
```

```python
# legend
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle='dashdot',marker='D', markersize=10, label='Virat')   #dashed
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'red', linestyle='dotted', linewidth=3, marker='o', label='Virat')
plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
plt.legend(loc = 'best')   # essential for showing the label

# loc possible values, by default loc = 'best'
# 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
```

```python
price = [54000,55000,56000,57000,49000,7900000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year, price)


# bcz of the last value being so big the previous curve looks almost flat
# now if we want we cam trim the graph
plt.ylim(0, 60000)
```

```python
# grid

plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle='dashdot',marker='D', markersize=10, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'red', linestyle='dotted', linewidth=3, marker='o', label='Virat')
plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
plt.legend(loc = 'best')
plt.grid()
```

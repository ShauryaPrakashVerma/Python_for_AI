### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical

```python
# bar chart
# x --> categorical
# y --> numerical

y = [10, 20, 30, 10, 50]
x = ['Red', 'Blue', 'Green', 'Yellow', 'Pink']

# plt.bar(x , y, color=['red', 'blue', 'green'])   --if colors are less then they will loop
plt.bar(x , y, color=['red', 'blue', 'green', 'yellow', 'pink'])
```

```python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
# plt.bar(df['batsman'], df['2015'])
# plt.bar(df['batsman'], df['2016'])
# this will cause the graph to overlap

plt.bar(np.arange(df.shape[0]), df['2016'])
```

```python
# to plot multiple bar charts side by side , there is an indirect method

plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)     # shifting to the left by 0.2
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)           # no shifting is done
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)     # shifting to the left by 0.2

plt.xticks(np.arange(df.shape[0]), df['batsman'])    # show the second label instead of the first label that is currently existing

plt.show()
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')   # without plt.xticks the names overlap each other
plt.xticks(rotation='vertical')

plt.show()
```

```python
# stacked bar chart
plt.bar(df['batsman'], df['2015'], label='2015')
plt.bar(df['batsman'], df['2016'], bottom=df['2015'], label='2016')
plt.bar(df['batsman'], df['2017'], bottom=(df['2015']+df['2016']), label='2017')

plt.legend()
plt.show()

```

### Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
# wanto assign color to a particular species in the dataset
# so to do this we assign a number to the species.
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
iris.sample(5)
```

```python
# cmap is basically a theme
# c basically colors similar points on basis of its value
# alpha is basically opacity  , lies between the range 0 - 1


plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap="winter", alpha=1)
plt.xlabel('Sepal length')
plt.ylabel('Petal length')

plt.colorbar()   # adds a color bar to the right

plt.show()
```

```python
# what if we want to increase the graph size

# figsize accepts a tuple, where 0 --> width, 1 --> length

plt.figure(figsize=(15,7))

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap="winter", alpha=1)
plt.xlabel('Sepal length')
plt.ylabel('Petal length')

plt.colorbar()   # adds a color bar to the right

plt.show()
```

```python
# annotations
# random_state is basically an attribute to lock the randomness, anyone using the same random_state value will get the same random sample value


batters = pd.read_csv('batter (1).csv')
sample = batters.head(100).sample(25, random_state=5)
sample
```

```python
plt.figure(figsize=(15,10))

plt.scatter(sample['avg'], sample['strike_rate'], s=sample['runs'])
for i in range(sample.shape[0]):
  plt.text(sample['avg'].values[i], sample['strike_rate'].values[i], sample['batter'].values[i])


# but in this graph we cant understand which dot is which graph
# we can do this with annotations
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4', fontdict={'size':12,'color':'brown'})

# we can also change font, change color, and many more font control settings
```

# Horizontal and Vertical lines

```python
# suppose we want to focus on only those points whose strike rate is greater than 130
plt.figure(figsize=(15,10))

plt.scatter(sample['avg'], sample['strike_rate'], s=sample['runs'])


# we can draw multiplelines
plt.axhline(130, color = "red")  # people above this line has sr above 130
plt.axhline(140, color = "yellow")   # people above this line have sr above 140
plt.axvline(30, color="green")   # people to the right have avg > 30

for i in range(sample.shape[0]):
  plt.text(sample['avg'].values[i], sample['strike_rate'].values[i], sample['batter'].values[i])
```

### Exceptions

If things go wrong during the execution of the program(runtime). It generally happens when something unforeseen has happened.

- Exceptions are raised by python runtime
- You have to takle is on the fly

#### **Examples**

- Memory overflow
- Divide by 0 -> logical error
- Database error

```python
# catching specific exception
try:
  m=5
  f = open('sample1.txt','r')
  print(f.read())
  print(m)
  print(5/2)
  L = [1,2,3]
  L[100]
except FileNotFoundError:
  print('file not found')
except NameError:
  print('variable not defined')
except ZeroDivisionError:
  print("can't divide by 0")
except Exception as e:
  print(e)
```

```python
# else  --> executes if the code in the try block does not raise an exception
      # --> if the block inside try raises an error then the else block is not triggered, then the except block is trigerred
try:
  f = open('sample1.txt','r')
except FileNotFoundError:
  print('file nai mili')
except Exception:
  print('kuch to lafda hai')
else:
  print(f.read())
```

```python
# finally   --> this block always executes regardless of whether the code in try block raises an exception or not
          # --> generally used to close connection to file, eg bluetooth, database etc.

try:
  f = open('sample1.txt','r')
except FileNotFoundError:
  print('file nai mili')
except Exception:
  print('kuch to lafda hai')
else:
  print(f.read())
finally:
  print('ye to print hoga hi')
```

```python
class Bank:

  def __init__(self, bal) -> None:
    self.bal = bal

  def withdraw(self,amount):
    if(amount<0):
      raise Exception('amount cant be negative')
    if(self.bal < amount):
      raise Exception("Not enough money")
    self.bal = self.bal - amount

obj = Bank(1000)
try:
  obj.withdraw(5000)
except Exception as e:
  print(e)
else:
  print(obj.bal)
```

```python
class SecurityError(Exception):

  def __init__(self,message):
    print(message)

  def logout(self):
    print('logout')

class Google:

  def __init__(self,name,email,password,device):
    self.name = name
    self.email = email
    self.password = password
    self.device = device

  def login(self,email,password,device):
    if device != self.device:
      raise SecurityError('bhai teri to lag gayi')
    if email == self.email and password == self.password:
      print('welcome')
    else:
      print('login error')


obj = Google('shaurya','shauryaverma22@gmail.com','1234','android')

try:
  obj.login('shauryaverma22@gmail.com','1234','windows')
except SecurityError as e:
  e.logout()
else:
  print(obj.name)
finally:
  print('database connection closed')
```

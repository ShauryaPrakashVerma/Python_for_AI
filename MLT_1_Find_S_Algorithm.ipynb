```python
import pandas as pd
import numpy as np
data = pd.read_csv("PlayTennis.csv")
```

```python
attributes = np.array(data)[:, :-1]
target = np.array(data)[:, -1]
```

```python
def train(attributes, target):
  specific_hypothesis = None
  for i, val in enumerate(target):
    if val == "yes":
      specific_hypothesis = attributes[i].copy()
      break

  if specific_hypothesis is None:
      raise ValueError("No positive examples found in the target to initialize the hypothesis.")

  for i, val in enumerate(attributes):
    if target[i] == "yes":
      for x in range(len(specific_hypothesis)):
        if val[x] != specific_hypothesis[x]:
          specific_hypothesis[x] = '?'
        else:
          pass

  return specific_hypothesis
```

```python
print("The final hypothesis is:", train(attributes, target))
print("Shaurya Prakash Verma")
print("2300971530120")
```

```python
class MeraList:

  def __init__(self):
    self.size = 1
    self.n = 0
    # create a c type array with size = self.size
    self.A = self.__make_array(self.size)


  def __len__(self):
    return self.n


  def append(self, item):
    if self.n == self.size:
      #resie
      self.__resize(self.size*2)
    #append
    self.A[self.n] = item
    self.n = self.n +1


  def pop(self):
    if self.n == 0:
      return "Empty List"

    print(self.A[self.n - 1])
    self.n = self.n - 1


  def clear(self):
    self.size = 1
    self.n = 0


  def __resize(self, new_capacity):
    # create new arry with new capacity
    B = self.__make_array(new_capacity)
    self.size = new_capacity
    # copy the content of old array to new array
    for i in range(self.n):
      B[i] = self.A[i]
    # reassign a
    self.A = B

  def __str__(self):
    result = ""
    for i in range(self.n):
      result = result + str(self.A[i]) + ","

    return '[' + result[:-1] + ']'


  def __getitem__(self,index):
    if 0<=index <self.n:
      return self.A[index]
    else:
      return "IndexError - Index out of bound"

  def find(self, item):
    for i in range(self.n):
      if self.A[i] == item:
        return i
    return "ValueError - Not in list"


  def insert(self, pos, item):
    if self.n == self.size:
      self.__resize(self.size*2)


    for i in range(self.n, pos, -1):
      self.A[i] = self.A[i-1]

    self.A[pos] = item
    self.n = self.n + 1

  def __delitem__(self,pos):
    if 0 <= pos < self.n :
      for i in range(pos, self.n-2,1):
        self.A[1] = self.A[i+1]

      self.n = self.n - 1

  def remove(self, item):
    pos = self.find(item)

    if type(pos) == int:
      self.__delitem__(pos)
    else:
      return pos



  def __make_array(self, capacity):
    # creates a C type array(static array, referential array) with size capacity
    return (capacity*ctypes.py_object)()

```

```python
from ast import Delete
class LinkedList:

  def __init__(self):
    # empty linked list
    self.head = None
    self.n = 0



  # length pf the linked list
  def __len__(self):
    return self.n


  # insert from head
  def insert_head(self, value):

    # new node
    new_node = Node(value)

    # create
    new_node.next = self.head
    self.head = new_node
    self.n = self.n + 1


  # insert from tail
  def append(self, value):

    new_node = Node(value)

    if self.head == None:
      self.head = new_node
      self.n = self.n + 1
      return

    current = self.head
    while(current.next != None):
      current = current.next

    current.next = new_node
    self.n = self.n + 1


  def insert_after(self, value, after):
    new_node = Node(value)

    current = self.head
    while current != None:
      if current.data == after:
        break
      current = current.next

    if current != None:
      new_node.next = current.next
      current.next = new_node
      self.n = self.n + 1
    else:
      return "Item not found"


  # clear the Linked list
  def clear(self):
    self.head = None
    self.n = 0


  # remove from the front
  def delete_head(self):
    if self.head == None:
      return "Empty linked list"
    self.head = self.head.next
    self.n = self.n - 1

  # remove from tail (pop)
  def pop(self):

    current = self.head

    if current == None:
      return "Empty linked list"

    if current.next == None:
      self.delete_head()
      self.n = self.n -1
      return

    while current.next.next != None:
      current = current.next

    current.next = None
    self.n = self.n -1


  # remove from the middle
  def remove_middle(self, value):
    if self.head == None:
      return "Empty Linked list"

    if self.head.data == value:
      self.delete_head()
      self.n = self.n - 1
      return

    current = self.head
    while current.next != None:
      if current.next.data == value:
        break
      current = current.next

    if current.next == None:
      return "Not found"
    else:
      current.next = current.next.next
      self.n = self.n - 1


  def search_value(self, value):
    current = self.head
    count = 0

    while current != None:
      if current.data == value:
        return count
      current = current.next
      count = count + 1
    else:
      return "Item not found"


  # item at particular index item
  def __getitem__(self, index):

    current = self.head
    pos = 0

    while current != None:
      if pos == index:
        return current.data
      current = current.next
      pos = pos + 1
    else:
      return "IndexError"

# Write a pyhton program to find the maximum value in a linked list and replace it with a given value.
# assume that the linked list is populated with whole numbers and there is only one max number

  def max_num(self, value):

    current = self.head
    max = current
    while(current.next != None):  # can also be current != None
      if current.data > max.data:
        max = current
      current = current.next

    max.data = value


  # sum of numbers at odd position in liked list
  def odd_sum(self):
    count = 0
    sum = 0
    current = self.head
    while current != None:
      if count %2 != 0:
        sum += current.data
      current = current.next
      count += 1

    print(sum)


  # reverse a linked list containing integer data. without creating a new linked list (in place reversal)
  # for eg 1->2->3->4 becomes 4->3->2->1

  def reverse(self):

    prev_node = None
    curr_node = self.head

    while curr_node != None:
      next_node = curr_node.next
      curr_node.next = prev_node

      prev_node = curr_node
      curr_node = next_node

    self.head = prev_node


  def __str__(self):

    current = self.head

    result = ""

    while current != None:
      result = result + str(current.data) + "->"
      current = current.next

    return result[:-2]
```

# Stack implementation using Linked List

```python
class Node:

  def __init__(self, value):
    self.data = value
    self.next = None
```

```python

class Stack():

  def __init__(self):
      self.top = None

  def isEmpty(self):
    return self.top == None

  def push(self, value):
    new_node = Node(value)
    new_node.next = self.top

    self.top = new_node

  def peek(self):
    if self.top == None:
      return "Stack Empty"
    return self.top.data

  def pop(self):
    if self.top == None:
      return "Stack Empty"
    data =  self.top.data
    self.top = self.top.next
    return data


  def traversal(self):
    temp = self.top

    while temp != None:
      print(temp.data)
      temp = temp.next

```

```python
# reverse a string using a stack. although this is not appropriate method to do so, since the size of stack increases as the size of the string increases.

def reverse_string(test_str):

  s = Stack()

  for char in test_str:
    s.push(char)

  result = ""
  while(not s.isEmpty()):
    result = result + s.pop()

  print(result)
```

```python
# Text Editor Problem

# input string: Hello
# pattern: uuur (u- Undo, R- redo)

# problem - when tring to redo a complete stack it appends the "stack Empty" string to the result


def text_editor(string, pattern):

  r = Stack()
  u = Stack()

  # whole text to be pushed in redo (r) stack
  for char in string:
    r.push(char)

  # peror
  for char in pattern:
    if char == "u":
      u.push(r.pop())
    else:
      r.push(u.pop())

  # printing the final string:
  res = ""
  while(not r.isEmpty()):
    res = r.pop() + res

  print(res)
```

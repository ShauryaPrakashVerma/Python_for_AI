```python
def xor_operation(x, y):
    return 0 if x == y else 1
```

```python
def dwd(old_data, divisor):
    cal = [0] * len(divisor)
    data = old_data + [0] * len(divisor)

    # copy first part
    for i in range(len(divisor)):
        cal[i] = data[i]

    # division process
    for i in range(len(old_data)):
        print(f"{i+1}.) First data bit is : {cal[0]}")
        print("Remainder : ", end="")

        if cal[0] == 1:
            for j in range(1, len(divisor)):
                cal[j-1] = xor_operation(cal[j], divisor[j])
                print(cal[j-1], end="")
        else:
            for j in range(1, len(divisor)):
                cal[j-1] = xor_operation(cal[j], 0)
                print(cal[j-1], end="")

        # bring next bit
        cal[len(divisor)-1] = data[i + len(divisor)]
        print(cal[len(divisor)-1])

    return cal
```

```python
def receive_data(data, divisor):
    rem = dwd(data, divisor)

    for bit in rem:
        if bit != 0:
            print("Currupted data received...")
            return

    print("Data received without any error.")
```
